```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('fortune1000.csv', index_col = 'Rank')
```

```python
df.head(2)
```

```python
df.info()
```

```python
df['State'] = df['Location'].apply(lambda x : x.split(',')[1])
```

```python
df.head()
```

# Top 5 States by Company Location

```python
df['State'].value_counts().head().plot(kind = 'bar', color = 'orange')
plt.title('Company Headquarters - State (Top 5 )', fontdict = {'fontweight' : 'bold', 'fontsize': '10' })
plt.xlabel('State')
plt.ylabel('No. of Companies')
```

# Profits by Sector

```python
sectors = df.groupby('Sector')
```

```python
sectors.agg({'Profits' : 'sum'}).plot(kind = 'bar')
plt.title('Total Profit by Sector', fontdict = {'fontweight': 'bold','fontsize': 12})
plt.xlabel('Sector')
plt.ylabel('Total Profit')
```

# Revenue by Sector

```python
sectors.agg({'Revenue': 'sum'}).plot(kind = 'bar', color = 'grey')
plt.title('Revenue by Sector', fontdict = {'fontweight': 'bold','fontsize': 12})
plt.xlabel('Sector')
plt.ylabel('Revenue')
```

# Total Companies in Each Sector

```python
plt.figure(figsize = (18,10))
sns.countplot(x = 'Sector', data = df)
plt.title('Total Companies in Each Sector', fontdict = {'fontweight':'bold', 'fontsize' : 18})
```

# Profits by Industry

```python
df1 = df.groupby('Industry')
```

```python
df1.agg({'Profits':'sum'}).sort_values('Profits', ascending = False).head(10).plot(kind = 'bar', color = 'slateblue')
plt.xlabel('Industry')
plt.ylabel('Profits')
plt.title('Profits by Industry')
```

# Revenue by Industry

```python
df1.agg({'Revenue':'sum'}).sort_values('Revenue', ascending = False).head(10).plot(kind = 'bar', color = 'teal')
plt.xlabel('Industry')
plt.ylabel('Revenue')
plt.title('Revenue by Industry')
```

# Number of Companies by Each Industry

```python
df['Industry'].value_counts().head(10).plot(kind = 'bar',color = 'darkkhaki')
plt.title('Total Companies by Each Industry')
plt.xlabel('Industry')
plt.ylabel('Total Companies')
```

# Interpretations

- Most of the Top 100 Companies have headquarters in California, Texas and New York.
- Financial Sector had the biggest Profits and Revenue out of all the Sectors.
- Financial Sector had the most number of Companies.
- Industry which contributed the maximum profits is Commercial Banks whereas maximum revenue was for Petroleum Refining.
- Apple had the maximum profits out of all the companies.
